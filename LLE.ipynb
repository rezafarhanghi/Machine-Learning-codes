```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.manifold import LocallyLinearEmbedding as LLE
%matplotlib inline
```

```python
n_samples = 1500
```

```python
roll = datasets.make_swiss_roll(n_samples=n_samples, noise=0.05)[0]

# fig = plt.figure()
# ax = fig.add_subplot(111, projection='3d')
# ax.plot(roll[:, 0], roll[:, 1], roll[:, 2], 'ko')
plt.plot(roll[:, 0], roll[:, 2], 'k.')
```

```python
s_curve = datasets.make_s_curve(n_samples=n_samples, noise=0.05)[0]

plt.plot(s_curve[:, 0], s_curve[:, 2], 'k.')
```

```python
blobs = datasets.make_blobs(n_samples=n_samples, n_features=3, random_state=8)[0]
transformation = [[0.2, -0.3, 5], [0.6, -0.6, 1], [-0.4, 0.8, 3]]
blobs = np.dot(blobs, transformation)

# fig = plt.figure()
# ax = fig.add_subplot(111, projection='3d')
# ax.plot(blobs[:, 0], blobs[:, 1], blobs[:, 2], 'ko')
plt.plot(blobs[:, 0], blobs[:, 2], 'k.')
```

```python
dr = LLE(n_components=2)
X = dr.fit_transform(roll)

plt.plot(X[:, 0], X[:, 1], 'b.')
```

```python
dr = LLE(n_components=2)
X = dr.fit_transform(s_curve)

plt.plot(X[:, 0], X[:, 1], 'b.')
```

```python
dr = LLE(n_components=2)
X = dr.fit_transform(blobs)

plt.plot(X[:, 0], X[:, 1], 'b.')
```
